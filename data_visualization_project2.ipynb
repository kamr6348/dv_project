```python
# integrating medals.xlsx with the data set
new_dataset = pd.merge (reg, medals,on=['Team'])
new_dataset
```

```python
# feature engineering 

#converting height from cm to meter =>  #new_dataset['Height'] = new_dataset['Height'] /100
# calculating BMI
new_dataset['BMI'] = (new_dataset['Weight'])/(new_dataset['Height']**2)
# calculating obesity indicator
new_dataset['Obesity Indicator'] = (new_dataset['BMI'] >= 25)*1


new_dataset
```

```python
# research question for feature engineering 
# what is the number of overweighted participants by using BMI ?

overweight_BMI = new_dataset[(new_dataset['BMI']>25)]
counter_overweight_BMI = new_dataset.query("BMI>25 ").shape[0] 
print("the number of overwight is",counter_overweight_BMI)
```

```python
# DATA CLEANING FILLING THE MISSING VALUES AGE,WEIGHT AND HEIGHT WITH THE AVERAGE
# AGE
avg_age = athlete['Age'].mean()
print("average age",avg_age)
athlete['Age'].fillna(avg_age)
# what is the average age of partcipants ?
```

```python
# WEIGHT
avg_weight = athlete['Weight'].mean()
print("average weight",avg_weight)
athlete['Weight'].fillna(avg_weight)
# what is the average weight of partcipants ?
```

```python
# HEIGHT
avg_height = athlete['Height'].mean()
print("average height",avg_height )
athlete['Height'].fillna(avg_height)
# what is the average height of partcipants ?
```

```python
# Data cleaning handling outliers
# handling height
Q1 = athlete['Height'].quantile(0.25)
Q3 = athlete['Height'].quantile(0.75)
IQR = Q3 - Q1
arr_1 = athlete['Height'] < (Q1 - 1.5 * IQR)
arr_2 = athlete['Height'] > (Q3 + 1.5 * IQR)
mask_height = (arr_1 | arr_2)
athlete_height_handled = athlete[~(mask_height)]

# to see the number of outliers uncomment the next line
#print("No. of outliers: ", mask_height.sum())
athlete_height_handled
```

```python
# Data cleaning handling outliers
# handling weigth
Q1 = athlete['Weight'].quantile(0.25)
Q3 = athlete['Weight'].quantile(0.75)
IQR = Q3 - Q1
arr_1 = athlete['Weight'] < (Q1 - 1.5 * IQR)
arr_2 = athlete['Weight'] > (Q3 + 1.5 * IQR)
mask_weight = (arr_1 | arr_2)
athlete_weight_handled = athlete[~(mask_weight)]

# to see the number of outliers uncomment the next line
#print("No. of outliers: ", mask_weight.sum())
athlete_weight_handled
```
